```python
import numpy as np
months=np.array(["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
sales=[]
print("Enter the sales for each month (in $1000 ):")
for month in months:
  values= float(input(f"Enter the sales for {month}: "))
  sales.append(values)
```

```python
sales = np.array(sales)
print("total sales of the year:", np.sum(sales))
print("Average sales of the year:", np.mean(sales))
print("Maximum sales of the year:", np.max(sales))
print("Minimum sales of the year:", np.min(sales))
```

```python
best_month = months[np.argmax(sales)]
worst_month = months[np.argmin(sales)]
print("Best month for sales:", best_month)
print("Worst month for sales:", worst_month)
```

```python
above_avg=months[sales>np.mean(sales)]
below_avg=months[sales<np.mean(sales)]
print("Months with sales above average:", above_avg)
print("Months with sales below average:",below_avg)
```
